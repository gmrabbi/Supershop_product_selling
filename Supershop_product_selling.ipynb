```python
# Loop through each fruit and create a bar plot
for fruit in Fruits_list:
    # Filter the DataFrame to include only rows where the specific fruit is present in the 'Fruits' column
    filtered_df = df[df['Fruits'].str.contains(fruit)]
    
    # Group by 'Country' and count occurrences
    fruit_count = filtered_df.groupby('Country', as_index=False)[fruit].count()
    
    # Sort the result
    fruit_count = fruit_count.sort_values(by=fruit, ascending=False)
    
    # Plot the data
    fig = px.bar(data_frame=fruit_count, x='Country', y=fruit, title=f'Country vs {fruit} Count', color='Country')
    
    fig.update_xaxes(tickangle=90)
    fig.update_layout(showlegend=False)
    fig.show()
```

<h5>This function is for dreace dictionary of country and price frequency </h5>
<p>if the price is gratter then stored that one, than the price will be overrite otherwise not</p>

```python
def country_freq_price(product):
    country_price = dict()
    for x in df.index:
        country = df.loc[x, "Country"]
        price = df.loc[x, product]

        if (country not in country_price):
            country_price[country] = price
        else:
            if(price > country_price[country]) :
                country_price[country] = price

    sorted_country_price = dict(sorted(country_price.items(), key=lambda item: item[1], reverse=True))
    return sorted_country_price
```

<h4>Country VS Height price </h4>
<p>This will plot the countries where the product is sold at the highest price.</p>

```python
for fruit in Fruits_list:
    data = country_freq_price(fruit)

    df1 = pd.DataFrame(list(data.items()), columns=['Country', 'Value'])

    # Plot the data using Plotly Express
    fig = px.bar(df1, x='Country', y='Value', title='Country vs Height price of '+fruit, labels={'Value': 'Price'}, template='plotly_white')

    # Update the layout for better visualization
    fig.update_layout(xaxis_title='Country', yaxis_title='Value', xaxis_tickangle=-90 )
    fig.show()
```

<p>The Counter function for count the Male and Female for each fruit</p>

```python
def male_female_counter(product):
    male_count =0
    female_count = 0

    count = 0

    for x in df.index:
        if( (product in df.loc[x, "Fruits"]) ):
            count += 1
            if((df.loc[x, "Gender"] == "Male")):
                male_count += 1
            elif((df.loc[x, "Gender"] == "Female")):
                female_count += 1
    return [count, male_count, female_count]
```

<p>Month VS Frequency of the month for invididual product</p>
<p>The Graph Represent, in which month which produch product sell(Most/least)</p>

```python
# Convert the 'Date' column to datetime format and extract the month
df['Date'] = pd.to_datetime(df['Date'], format='%d %b, %Y')
df['Month'] = df['Date'].dt.strftime('%b')

# Loop through each fruit and create a bar plot
for fruit in Fruits_list:
    # Filter rows where the specific fruit is present in the 'Fruits' column
    filtered_df = df[df['Fruits'].str.contains(fruit)]
    
    # Aggregate the data: Group by month and count occurrences of the specific fruit
    aggregated_data = filtered_df.groupby('Month')[fruit].count().reindex(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    
    # Create a bar plot using Plotly Express
    fig = px.bar(
        x=aggregated_data.index, 
        y=aggregated_data.values, 
        title=f'Monthly Frequency of {fruit}', 
        labels={'x': 'Month', 'y': 'Total Quantity'},
        text=aggregated_data.values
    )
    
    # Update the layout for better readability
    fig.update_xaxes(tickangle=90)
    fig.update_layout(showlegend=False)
    
    # Show the plot
    fig.show()

# Clean up: Drop the temporary 'Month' column
df.drop(columns="Month", inplace=True)
```

<h4>This is our main part , we save the data item by the group of fruits</h4>
<p>we do this, because , suppose we offer/discount on a product, it that time we can notify/inform them through their mail/Phone number</p>
<p>Save the data items of individual fruits in Month directory and the file name will be the fruit name</p>

```python
import os

# Ensure the directory exists
output_dir = ".\\Fruit\\"
os.makedirs(output_dir, exist_ok=True)

fruit_dfs = {}

# Create individual DataFrames for each fruit
for fruit in Fruits_list:
    # Create a new DataFrame with only the relevant columns
    fruit_df = dfc[['Customer ID', 'Date', 'Time', 'First Name', 'Last Name', 'Gender', 'Gmail Address', 'Phone Number', 'Country', 'Contact Later', 'Fruits', fruit,
                   'Total', 'state_address', 'street', 'Zip_Code']].copy()
    
    # Filter rows where the 'Fruits' column contains exactly the specific fruit
    fruit_df = fruit_df[fruit_df['Fruits'].str.split(',').apply(lambda x: fruit.capitalize() in [f.strip() for f in x])]
    
    # Store the DataFrame in the dictionary
    fruit_dfs[fruit] = fruit_df
    
    # Save the DataFrame to a CSV file
    file_path = os.path.join(output_dir, f"{fruit}.csv")
    fruit_df.to_csv(file_path, index=False)
    print(f"Saved DataFrame for {fruit} to {file_path}")
```
